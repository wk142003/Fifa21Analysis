### Converting Value and Wage into Integer Data Type

```python
# To convert the Value Column, keep the current Value column and create a new Value Int column
updated_df1['Value_int'] = updated_df1["Value"].astype("str") # stores a string version of the items in Value
updated_df1['Value_int'] = updated_df1["Value"].str.replace('€','') #.str is to access string functions

updated_df1['Wage_amt'] = updated_df1["Wage"].astype("str")
updated_df1['Wage_amt'] = updated_df1["Wage"].str.replace('€','')

def convert_amount(value):
    if 'M' in value:
        return float(value.replace('M', '')) * 1e6
    elif 'K' in value:
        return float(value.replace('K', '')) * 1e3
    else:
        return float(value)

updated_df1['Value_int'] = updated_df1["Value_int"].apply(convert_amount)
updated_df1['Wage_amt'] = updated_df1["Wage_amt"].apply(convert_amount)
```

### Classify the players by position, ATT, MID, DEF or GK

```python
def position(best_pos):
    if best_pos in ["RW", "LW", "CF", "ST"]:
        return "ATT"
    elif best_pos in ["CAM", "CDM", "CM", "LM", "RM"]:
        return "MID"
    elif best_pos in ["CB", "LB", "RB", "RWB", "LWB"]:
        return "DEF"
    elif best_pos == "GK":
        return "GK"

updated_df1["Field Position"] = updated_df1["Best Position"].apply(position)
```
